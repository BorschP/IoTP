```python
import os

from ultralytics import YOLO
import cv2
import numpy as np



cap = cv2.VideoCapture("C:/Users/Borsch/Desktop/PIoT/enFire.mp4")
# cap = cv2.VideoCapture(0)
# cap = cv2.VideoCapture("C:/Users/Borsch/Desktop/5555.jpg")

model_path = os.path.join("C:/Users/Borsch/Desktop/PIoT_V2/runs/detect/train/weights/last.pt")#detect fire
model_path2 = os.path.join("C:/Users/Borsch/Desktop/PIoT/runs/detect/train2/weights/best.pt")#detect smoke

# Load a model
model = YOLO(model_path)  # load a custom model
model2 = YOLO(model_path2) # load a custom model

threshold = 0.2

# classes = ["0" , "smoke"]
class_name_dict = {0: 'smoke' ,1: 'zfire'}

# color = np.random.uniform(0,100, size=(len(class_name_dict),3))


while True:

        ret,frame = cap.read()  
        if ret :      
            results = model(frame)[0]#ทำให้วิดีโอเล่นช้าลง
            results2 = model2(frame)[0]
            # print(results)
            for result in results.boxes.data.tolist():
                # print("check")
                startX, startY, endX, endY, score, class_id = result
                # print("1 : ",startX)
                # print("2 : ",startY)
                # print("3 : ",endX)
                # print("4 : ",endY)
                # print("5 : ",score)
                # print("6 : ",class_id)
                if score > threshold:
                    cv2.rectangle(frame , (int(startX) , int(startY)) , (int(endX) , int(endY)) , (0,255,0) , 2)
                    cv2.putText(frame , class_name_dict[class_id].upper(),(int(startX), int(startY)) , cv2.FONT_HERSHEY_DUPLEX , 0.6 , (255,255,255) , 1 )
            
            for result in results2.boxes.data.tolist():
                 startX1, startY1, endX1, endY1, score, class_id = result
                 if score > threshold:
                    cv2.rectangle(frame , (int(startX1) , int(startY1)) , (int(endX1) , int(endY1)) , (0,255,0) , 2)
                    cv2.putText(frame , class_name_dict[class_id].upper(),(int(startX1), int(startY1)) , cv2.FONT_HERSHEY_DUPLEX , 0.6 , (255,255,255) , 1 )
                 
        
        
        
        
        cv2.imshow("Frame",frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        # print(frame.shape)

cap.release()
cv2.destroyAllWindows()



```
